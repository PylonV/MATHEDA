```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"..\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

# N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python

chart = chip_count.plot(kind='bar',
                        color=['#126782', '#126782', '#126782', '#8C92AC', '#8C92AC'],
                        figsize=(10, 5))
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)',
          fontfamily='Arial',  
          fontsize=14,      
          fontweight='bold',
          color='#4A4A4A')    
plt.xlabel('', fontsize=12, fontfamily='Arial', color='#4A4A4A')  
plt.ylabel('N Chips', fontsize=12, fontfamily='Arial', color='#4A4A4A')  
chart.set_xticklabels(chip_count.index, rotation=0, fontsize=10, fontfamily='Arial', color='#4A4A4A') 
chart.tick_params(axis='y', labelsize=10, colors='#4A4A4A')  

plt.show()
```

# N CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

```python

sns.set(style="whitegrid")
chart = cpu_count.plot(kind='bar', figsize=(6, 3), color='#4C72B0', edgecolor='black')

plt.title('CPU Manufactured',
          fontfamily='Arial',  
          fontsize=16,       
          fontweight='bold',  
          color='#333333')   

plt.ylabel('N Count',
           fontsize=12,
           fontfamily='Arial',
           color='#333333')  
plt.xlabel('',
           fontsize=12,
           fontfamily='Arial',
           color='#333333')  
chart.set_xticklabels(cpu_count.index, rotation=0, fontsize=10, fontfamily='Arial', color='#333333')  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() 
```

# Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp = df_tdp.sort_values(by='TDP')
df_tdp
```

```python
df_tdp = df_tdp.head(7)
df_tdp
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5), facecolor='#E7E3D2')
plt.bar(df_tdp['Product'], df_tdp['TDP'], color='#4C72B0', edgecolor='black')
plt.title('Products with the Lowest TDP',
          fontfamily='Arial',
          fontsize=16,       
          fontweight='bold',
          color='#333333')   
plt.ylabel('Watts',
           fontsize=12,
           fontfamily='Arial',
           color='#333333')  
plt.xticks(rotation=45, ha='right', fontsize=10, fontfamily='Arial', color='#333333')  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() 
plt.show()
```

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5), facecolor='#E7E3D2')
plt.plot(df_ps['Year'], df_ps['Process_Size'], marker='o', color='#4C72B0', linewidth=2, markersize=6)
plt.title('GPU Process Size Trend',
          fontfamily='Arial', 
          fontsize=16,        
          fontweight='bold', 
          color='#333333')    
plt.ylabel('nm',
           fontsize=12,
           fontfamily='Arial',
           color='#333333') 
plt.xlabel('Year',
           fontsize=12,
           fontfamily='Arial',
           color='#333333')  
plt.xticks(fontsize=10, fontfamily='Arial', color='#333333')  
plt.yticks(fontsize=10, fontfamily='Arial', color='#333333')  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() 
plt.show()
```
